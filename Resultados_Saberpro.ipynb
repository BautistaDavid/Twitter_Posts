```python
# Manejo de datos
import pandas as pd   

# Visualización de datos
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_excel('saberpro_data_.xlsx',)   # Se carga el archivo
df['Universidad'] = df['Universidad'].str.strip()  # Se modifica un error en variable Universidad
```

```python
# Se configura opciones de la caja del boxplot
PROPS = {'boxprops':{'edgecolor':'black'}, 'medianprops':{'color':'black'},'whiskerprops':{'color':'black'},'capprops':{'color':'black'}}

# Se configuta los labels del eje y
ylabels = ['Universidad\n Externado','Universidad\n Nacional','Universidad\n de los Andes',\
           'Universidad\n Javeriana','Universidad\n Santo Tomas']

# Se configura los colores de cada boxplot
paleta = {uni: "#9ad056" if uni == "Santo Tomas" else "#fff27f" for uni in df.Universidad.unique()}
# paleta = {'Externado':'#9ad056','Nacional':'#666666','Andes':'#fff27f','Pontificia  Javeriana':'#325a94','Santo Tomas':'#00488a'}

# Se construye cada elemento de la figura
plt.figure(figsize=(10,7))
plt.suptitle('Distribución Puntaje Global SaberPro\n 2019-II Pregrado en Economía',fontsize=20,weight='bold',x=0.5,y=1.02,ha='center')
sns.boxplot(data=df,y='Universidad',x='Puntaje Global',palette=paleta,**PROPS)
plt.yticks(ticks=[0,1,2,3,4],labels=ylabels,fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('')
plt.xlabel('')
plt.grid(axis='x',linestyle='dotted')
for i in ['bottom','left','top','right']:
  plt.gca().spines[i].set_linewidth(0)
plt.text(x=210,y=5.6,s='Autor: David Bautista\n          @PipeBau_',fontsize=14)
plt.savefig('boxplot_global.png',dpi=300, bbox_inches='tight')
plt.show()
```
